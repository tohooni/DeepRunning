```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np

#매직명령어 inline(정적인 시각화, notebook(interative한 시각화)
%matplotlib inline

# plt.type.available['여러가시 스타일 적용가능']
```

```python
x = np.linspace(0,10, 100)
```

```python
plt.plot(x, np.sin(x),"-", linewidth=1, label="sin(x)")
plt.plot(x, np.cos(x),"--",linewidth=1, label="cos(x)")
plt.xlim(0,10)
plt.ylim(-1,1)
plt.title("Sine & Cosine Curve")
plt.xlabel("x")
plt.legend() #범례
```

```python
#스타일 컨텍스트 매니저를 이용한 플롯 스타일의 일시적 설정변경
with plt.style.context("ggplot"):
    plt.plot(x, np.sin(x),"-", linewidth=1, label="sin(x)")
    plt.plot(x, np.cos(x),"--",linewidth=1, label="cos(x)")
    plt.xlim(0,10)
    plt.ylim(-1,1)
    plt.title("Sine & Cosine Curve")
    plt.xlabel("x")
    plt.legend() #범례
```

```python
#객체지향 인터페이스 Figure와 Axes갤체의 사용
fig = plt.figure() #캠버스
ax  = plt.axes() #축

x = np.linspace(0,10,100)
plt.plot(x, np.sin(x),"-", linewidth=1, label="sin(x)")
plt.plot(x, np.cos(x),"--",linewidth=1, label="cos(x)")
plt.xlim(0,10)
plt.ylim(-1,1)
plt.title("Sine & Cosine Curve")
plt.xlabel("x")
plt.legend() #범례
```

```python
#subplots함수를 이용해 여러개의 플롯을 그리기
fig, ax = plt.subplots(2,1)

ax[0].plot(x, np.sin(x), 'b-', linewidth=0.5, label="sin(x)")
ax[0].set_title("Sine & Cosizne Curve")
ax[0].set(xlim=[0,10], ylim=[-1,1])
ax[0].legend()

ax[1].plot(x, np.cos(x), 'g-', linewidth=0.5, label="cos(x)")
ax[1].set(xlim=[0,10], ylim=[-1,1])
ax[1].legend()
```

```python
#붓꽃 데이터 불러오기
from seaborn import load_dataset
iris = load_dataset("iris")
iris.info()
```

```python
iris.head()
```

```python
import pandas as pd

iris["species_cat"] = pd.Categorical(iris.species)
iris.info()
```

```python
iris.head()
```

```python
plt.scatter(iris.sepal_length, iris.sepal_width, c=iris.species_cat.cat.codes, alpha=0.2, cmap="viridis")
plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max())
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
```

```python
species = iris.species.unique()
species
```

```python
for sp in species:
    x = iris.loc[iris["species"] == sp, "sepal_length"]
    y = iris.loc[iris["species"] == sp, "sepal_width"]
    plt.scatter(x, y, label=sp, alpha = 0.2)

plt.title("iris")
plt.xlim(iris.sepal_length.min(), iris.sepal_length.max())
plt.ylim(iris.sepal_width.min(), iris.sepal_width.max())
plt.xlabel(iris.columns[0])
plt.ylabel(iris.columns[1])
plt.legend()
```

```python
tips = load_dataset('tips')
tips.info()
```

```python
plt.hist(x=tips.tip, rwidth=0.9, bins=20)
```

```python
import scipy as sp

bins = 20
plt.hist(x=tips.tip, rwidth=0.9, bins=bins, density=True)
plt.title("tips")
plt.xlabel("tip")
plt.ylabel('density')
```

```python
tips.mean(), tips.std()
```

```python
#박스플롯 4분위수
tips.loc[tips.sex=="Male", "tip"].describe().round(1)
```

```python
#박스플롯을 활용하여 이상치 발견하기
labels = []
tip_list = []

for label, df_per_sex in tips.groupby("sex"):
    labels.append(label)
    tip_list.append(df_per_sex["tip"].tolist())

print(labels, tip_list)    
    
plt.boxplot(tip_list, labels=labels)
plt.title("tips")
plt.xlabel("sex")
plt.ylabel("tip")
```
